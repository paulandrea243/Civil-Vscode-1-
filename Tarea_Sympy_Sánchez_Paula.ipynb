# 2. Introducción Teórica Concisa

Sympy es una biblioteca de Python para matemáticas simbólicas que incluye un módulo llamado `continuum_mechanics`, diseñado para el análisis estructural. Dentro de este módulo, `Arch` y `Truss` son herramientas poderosas para el estudio de elementos estructurales fundamentales.

*   **Módulo `Arch`**: Permite modelar y analizar arcos, estructuras curvas que transfieren cargas principalmente por compresión. Es útil para determinar las reacciones en los apoyos, diagramas de fuerzas internas (normal, cortante y momento) y deflexiones. Su utilidad radica en la capacidad de manejar geometrías complejas y diferentes tipos de carga para comprender el comportamiento de arcos en puentes, edificios y otras construcciones. Sin embargo, la clase `Arch` en Sympy está principalmente orientada a arcos de tres articulaciones o empotrados; para arcos biarticulados generales, a menudo se requiere un enfoque de ecuaciones de equilibrio explícitas.

*   **Módulo `Truss`**: Se utiliza para modelar y analizar cerchas (o armaduras), estructuras compuestas por miembros rectos conectados en sus extremos por uniones articuladas. Las cerchas son eficientes para cubrir grandes luces y soportar cargas, ya que sus miembros están sometidos principalmente a tracción o compresión axial. Este módulo facilita el cálculo de las fuerzas internas en cada barra y las reacciones en los apoyos, lo cual es crucial para el diseño seguro y económico de techos, puentes y torres.

## Problema 1 (Arch): Arco Biarticulado con Carga Puntual en la Clave

Vamos a modelar un arco circular biarticulado (apoyos articulados en ambos extremos) con una carga puntual aplicada en su punto más alto (clave). Calcularemos las reacciones.

Dado que la clase `Arch` de Sympy está más orientada a arcos de tres articulaciones, resolveremos este problema planteando explícitamente las ecuaciones de equilibrio estático para un arco biarticulado simétrico.

```
from sympy import symbols, Eq, solve, pi
#from sympy.physics.continuum_mechanics.arch import Arch # Se comenta esta importación por no ser adecuada para arco biarticulado general

# --- Definición de Símbolos y Parámetros ---
# R: Radio del arco
# L: Luz del arco (distancia horizontal entre apoyos)
# H: Flecha del arco (altura desde la línea de apoyos hasta la clave)
# P: Carga puntual aplicada en la clave

R, L, H, P_sym = symbols('R L H P') # 'P_sym' para evitar conflicto con P_val

# Para un arco biarticulado simétrico con carga P en la clave
# Se plantean las ecuaciones de equilibrio:
# 1. Suma de fuerzas verticales = 0
# 2. Suma de momentos en un apoyo = 0 (para hallar reacciones verticales)
# 3. Suma de momentos en la clave (o vértice) = 0 (para hallar reacciones horizontales)

# Definir valores numéricos para el ejemplo
L_val = 10.0  # Luz del arco en metros
H_val = 2.5   # Flecha del arco en metros
P_val = 5000.0 # Carga puntual en la clave en Newtons

# Reacciones en los apoyos (Ay, By = verticales; H_reaccion = horizontal)
# Debido a la simetría y la carga centrada:
# Ay = P/2
# By = P/2
# Para H_reaccion (reacción horizontal):
# Momento en la clave (punto más alto del arco) desde el apoyo izquierdo:
# M_clave = Ay * (L/2) - H_reaccion * H = 0
# H_reaccion = (Ay * L/2) / H

Ay_val = P_val / 2
By_val = P_val / 2
H_reaccion_val = (Ay_val * (L_val / 2)) / H_val

print(f"--- Resultados para el Arco Biarticulado ---")
print(f"Carga puntual aplicada (P): {P_val} N")
print(f"Luz del arco (L): {L_val} m")
print(f"Flecha del arco (H): {H_val} m")
print(f"Reacción vertical en el apoyo A (Ay): {Ay_val:.2f} N")
print(f"Reacción vertical en el apoyo B (By): {By_val:.2f} N")
print(f"Reacción horizontal en ambos apoyos (H_reaccion): {H_reaccion_val:.2f} N")

# Para los esfuerzos internos (Normal, Cortante, Momento) en cualquier sección,
# se necesitaría la ecuación de la curva del arco y plantear el equilibrio
# en una sección genérica, lo cual es más complejo y está fuera del alcance de este ejemplo simplificado.
```

## Problema 2 (Truss): Cercha de Tres Barras

Modelaremos una cercha simple de tres barras con una carga puntual en un nudo. Calcularemos las fuerzas internas en las barras y las reacciones en los apoyos utilizando las ecuaciones de equilibrio en cada nudo.

```
from sympy import symbols, Eq, solve
from sympy.physics.continuum_mechanics.truss import Truss # Aunque se importa, no se usará directamente para la solución
import math
import matplotlib.pyplot as plt

# --- Definición de Nudos, Barras, Apoyos y Cargas ---

# Coordenadas de los nudos (x, y)
nodos_coords = {
    'N0': (0, 0),       # Apoyo izquierdo
    'N1': (3, 0),       # Apoyo derecho
    'N2': (1.5, 2)      # Nudo superior central
}

# Definir las barras (conectividad entre nudos)
barras = [
    ('N0', 'N2'), # Barra izquierda inclinada
    ('N1', 'N2'), # Barra derecha inclinada
    ('N0', 'N1')  # Barra horizontal base
]

# Definir los apoyos
apoyos = {
    'N0': {'type': 'pinned'}, # Apoyo fijo (reacciones en X y Y)
    'N1': {'type': 'roller'}  # Apoyo de rodillo (reacción solo en Y)
}

# Definir las cargas
load_val = 1000 # Valor de la carga en N
cargas = {
    'N2': (0, -load_val) # Carga de 1000 N hacia abajo en el nudo N2 (Fx, Fy)
}

# --- Variables Simbólicas para Fuerzas Internas y Reacciones ---
F_01, F_02, F_12 = symbols('F_01 F_02 F_12') # Fuerzas internas en las barras
R_0x, R_0y, R_1y = symbols('R_0x R_0y R_1y') # Reacciones en los apoyos

# --- Geometría del Truss para Componentes de Fuerza ---

# Distancias y alturas
L_base = nodos_coords['N1'][0] - nodos_coords['N0'][0] # Luz de la base
H_truss = nodos_coords['N2'][1] - nodos_coords['N0'][1] # Altura de la cercha
L_half_base = nodos_coords['N2'][0] - nodos_coords['N0'][0] # Distancia horizontal de N0 a N2

# Cálculo de los ángulos para las barras inclinadas
# Barra N0-N2
angle_02 = math.atan2(H_truss, L_half_base)
cos_02 = math.cos(angle_02)
sin_02 = math.sin(angle_02)

# Barra N1-N2 (es simétrica a N0-N2 en este caso, se puede verificar)
# delta_x_12 = nodos_coords['N1'][0] - nodos_coords['N2'][0]
# angle_12 = math.atan2(H_truss, delta_x_12)
# cos_12 = math.cos(angle_12)
# sin_12 = math.sin(angle_12)
# Usaremos las componentes para N0-N2 y N1-N2, notando la simetría de magnitudes y signos.
# Por simetría, cos_02 y sin_02 son válidos para la magnitud de las componentes de F_12.

# --- Ecuaciones de Equilibrio en Cada Nudo ---

# Nudo N0 (Apoyo Fijo: R_0x, R_0y)
# Suma Fx = 0: R_0x + F_01 + F_02 * cos_02 = 0
# Suma Fy = 0: R_0y + F_02 * sin_02 = 0
equations_N0 = [
    Eq(R_0x + F_01 + F_02 * cos_02, 0),
    Eq(R_0y + F_02 * sin_02, 0)
]

# Nudo N1 (Apoyo de Rodillo: R_1y)
# Suma Fx = 0: -F_01 - F_12 * cos_02 = 0  (componente x de F_12 va a la izquierda)
# Suma Fy = 0: R_1y + F_12 * sin_02 = 0
equations_N1 = [
    Eq(-F_01 - F_12 * cos_02, 0),
    Eq(R_1y + F_12 * sin_02, 0)
]

# Nudo N2 (Carga Aplicada: (0, -load_val))
# Suma Fx = 0: -F_02 * cos_02 + F_12 * cos_02 = 0
# Suma Fy = 0: -F_02 * sin_02 - F_12 * sin_02 + cargas['N2'][1] = 0
equations_N2 = [
    Eq(-F_02 * cos_02 + F_12 * cos_02, 0),
    Eq(-F_02 * sin_02 - F_12 * sin_02 + cargas['N2'][1], 0)
]

# Combinar todas las ecuaciones para resolver el sistema
all_equations = equations_N0 + equations_N1 + equations_N2

# Variables a resolver
symbols_to_solve = (F_01, F_02, F_12, R_0x, R_0y, R_1y)

solutions = solve(all_equations, symbols_to_solve)

print("--- Resultados para la Cercha de Tres Barras ---")
print("Soluciones para las fuerzas en las barras y reacciones:")
for var, val in solutions.items():
    print(f"{var}: {val:.2f} N")

# Interpretación de los resultados:
# Fuerzas en barras (F_01, F_02, F_12):
#   - Valor positivo: la barra está en tracción (se estira).
#   - Valor negativo: la barra está en compresión (se acorta).
# Reacciones (R_0x, R_0y, R_1y):
#   - Valor positivo: la reacción actúa en la dirección positiva del eje.
#   - Valor negativo: la reacción actúa en la dirección negativa del eje.
```

```
print("\n--- Resumen de Resultados ---")

# Resultados del Arco Biarticulado
print("\nArco Biarticulado:")
print(f"  Carga puntual aplicada (P): {P_val} N")
print(f"  Luz del arco (L): {L_val} m")
print(f"  Flecha del arco (H): {H_val} m")
print(f"  Reacción vertical en el apoyo A (Ay): {Ay_val:.2f} N")
print(f"  Reacción vertical en el apoyo B (By): {By_val:.2f} N")
print(f"  Reacción horizontal en ambos apoyos (H_reaccion): {H_reaccion_val:.2f} N")

# Resultados de la Cercha de Tres Barras
print("\nCercha de Tres Barras:")
print(f"  Carga puntual en N2: {cargas['N2'][1]} N (hacia abajo)")

if solutions:
    print("  Fuerzas internas en las barras:")
    print(f"    Barra N0-N1 (F_01): {solutions[F_01]:.2f} N (axial)")
    print(f"    Barra N0-N2 (F_02): {solutions[F_02]:.2f} N (axial)")
    print(f"    Barra N1-N2 (F_12): {solutions[F_12]:.2f} N (axial)")
    print("  Reacciones en los apoyos:")
    print(f"    Reacción horizontal en N0 (R_0x): {solutions[R_0x]:.2f} N")
    print(f"    Reacción vertical en N0 (R_0y): {solutions[R_0y]:.2f} N")
    print(f"    Reacción vertical en N1 (R_1y): {solutions[R_1y]:.2f} N")
else:
    print("  No se encontraron soluciones para las fuerzas en las barras y reacciones de la cercha.")

# --- Visualización de la Cercha ---
print("\nVisualización esquemática de la Cercha:")

plt.figure(figsize=(6, 4))

# Dibujar nudos
for node_name, (x, y) in nodos_coords.items():
    plt.plot(x, y, 'o', color='blue', markersize=8)
    plt.text(x + 0.1, y + 0.1, node_name)

# Dibujar barras
for n1, n2 in barras:
    x1, y1 = nodos_coords[n1]
    x2, y2 = nodos_coords[n2]
    plt.plot([x1, x2], [y1, y2], 'k-', linewidth=2)

# Dibujar apoyos
# Apoyo fijo en N0
plt.plot(nodos_coords['N0'][0], nodos_coords['N0'][1], '^', color='green', markersize=12, label='Apoyo Fijo')
# Apoyo de rodillo en N1
plt.plot(nodos_coords['N1'][0], nodos_coords['N1'][1], 's', color='red', markersize=10, label='Apoyo Rodillo')

# Dibujar carga
load_node = 'N2'
load_x, load_y = nodos_coords[load_node]
# La flecha se dibuja desde el nudo en la dirección de la carga.
# La magnitud de la flecha se escala para visualización.
plt.arrow(load_x, load_y, 0, cargas[load_node][1] / 1000,
          head_width=0.2, head_length=0.3, fc='purple', ec='purple', label='Carga Puntual')

plt.title('Esquema de la Cercha de Tres Barras')
plt.xlabel('Eje X (m)')
plt.ylabel('Eje Y (m)')
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.gca().set_aspect('equal', adjustable='box') # Asegura que la escala sea igual en ambos ejes
plt.legend()
plt.show()
```

### Visualización del Arco Biarticulado
